```python
import torch
from torch import nn
from torch.autograd import Variable
```

```python
# Backpropogation from 4 video-lesson 
# https://www.youtube.com/watch?v=ma2KXWblllc&list=PLlMkM4tgfjnJ3I-dbhO9JTw7gNty6o_2m&index=5&t=0s
```

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
alpha = 0.01

w = Variable(torch.Tensor([1.0]),  requires_grad=True)
```

```python
# our model for the forward pass
def forward(x, w):
    return x * w


# Loss function
def loss(y_predicted, y):
    return (y_predicted - y) * (y_predicted - y)

def gradient(x, y, weight):
	return  2 * x * (x * weight - y)

def weightUpdate(w, x, y):
	return  w - alpha * gradient(x, y, w)

# List of weights/Mean square Error (Mse) for each input
w_list = []
mse_list = []

print("before prediction for x = 4: y = ", forward(4, w).item())

for i in range(100):
	print("ITERATION = ", i)
	print("weight = ", w.data)

	for x_value, y_value in zip(x_data, y_data):
		print("x0 = ", x_value)
		print("y0 = ", y_value)
		y_predicted = forward(x_value, w)
		print("y1 = ", y_predicted)

		loss1 = loss(y_predicted, y_value)
		loss1.backward()
		print("loss = ", loss1)

		w.data = w.data - alpha * w.grad.item() 
		print("New weigth = ", w.grad.data[0])

		w.grad.data.zero_()

print("final prediction for x = 4: y = ", forward(4, w).item())
```
